```python
import numpy as np
cat_cols=data.select_dtypes(include=['object']).columns
num_cols = data.select_dtypes(include=np.number).columns.tolist()
print("Categorical Variables:")
print(cat_cols)
print("Numerical Variables:")
print(num_cols)
```

```python
data.head()
```

```python
for col in num_cols:
    print(col)
    print('Skew :', round(data[col].skew(), 2))
    plt.figure(figsize = (15, 4))
    plt.subplot(1, 2, 1)
    data[col].hist(grid=False)
    plt.ylabel('count')
    plt.subplot(1, 2, 2)
    sb.boxplot(x=data[col])
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("dataset/HR Attrition.csv") 
df.head()
```

```python
plt.hist(df.Age,edgecolor='k')

plt.show()
```

```python
plt.figure(figsize=(15,11))
plt.scatter(df.TotalWorkingYears,df.MonthlyIncome,c='b',s=10)
plt.xlabel('Working Years')
plt.ylabel('Income')
plt.title('Scatter chart of Working Years vs Income')
plt.show()
```

```python
sb.catplot(x='EmployeeCount',data=df,hue='Department',kind='count')
plt.show()
```
